```python
# the usual imports for data analytics
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load test datasets: tips and flights
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
tips['total_cost'] = tips['tip'] + tips['total_bill']
```

```python
# get correlation matrix
correlations = tips.corr(numeric_only=True)
```

```python
# heatmaps are nice to see the value differences in a correlation matrix
sns.heatmap(correlations, cmap="coolwarm", annot=True)
```

```python
# let's see the flights data correlation matrix
sns.heatmap(flights.corr(), cmap="coolwarm", annot=True)
```

```python
fp = flights.pivot_table(index="month", columns="year", values="passengers")
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap = "magma")
```

```python
sns.heatmap(fp, cmap = "magma", linecolor="gray", linewidths=1)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, cmap = "coolwarm", standard_scale=1)
```
