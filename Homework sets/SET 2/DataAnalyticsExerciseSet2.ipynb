<li>Finally, get the amount of rows grouped by each of the new Income Group –field values and print them out.</li> 


```python
# code for the advanced extra task

# download a dataframe
loans = pd.read_csv('loans.csv')
loans
```

```python
# create a funciton for the income group
# I do not know why but I had to add "_"
# to "Annual Income" to make this function work
# I also dropped all NaN values 
# to get only numbers for the function
loans = loans.dropna()
def get_income_group(Annual_Income):
    if Annual_Income <= 25000:
        return '$25k or less'
    elif 25001 <= Annual_Income <= 50000:
        return '$25k - $50k'
    elif 50001 <= Annual_Income <= 100000:
        return '$50k - $100k'
    elif 100001 <= Annual_Income <= 200000:
        return '$100k - $200k'
    else:
        return '$200k+'
```

```python
# run the function to create the new column 'Income Group'
loans['Income Group'] = loans['Annual Income'].apply(get_income_group)
```

```python
# Group by 'Income Group' and count number of rows for each group
income_group_sort = loans.groupby('Income Group').size()
```
